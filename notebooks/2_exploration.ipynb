```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_pickle('featured_df.pkl')
```

```python
df.head(1)
```

```python
channels = ['ch_%02d' % i for i in range(1,17)]
```

```python
fig, axes = plt.subplots(ncols=1, nrows=16, figsize=(16,80))

for i, channel in enumerate(channels):
    axes[i].set_ylim(-800, 800)
    axes[i].plot(df.loc[4100, channel]);
```

```python
fig, axes = plt.subplots(ncols=1, nrows=16, figsize=(16,80))

for i, channel in enumerate(channels):
    axes[i].set_ylim(-800, 800)
    axes[i].plot(df.loc[500, channel]);
```
